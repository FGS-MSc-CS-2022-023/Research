```python
import numpy as np
import torch
```

```python
X = torch.tensor([1,2,3,4], dtype=torch.float32)
Y = torch.tensor([2,4,6,8], dtype=torch.float32)
```

```python
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
a = torch.tensor(0.0, requires_grad=True)
```

```python
def forward(X):
    return w * X
```

```python
def loss(y, y_pred):
    return ((y_pred - y)**2).mean()    
```

```python
for i in range(100):
    y_pred = forward(X)
    diff = loss(y_pred, Y)
    print('error', diff.item())
    diff.backward()
    with torch.no_grad():
        w -= 0.05 * w.grad
    print('\tgrad', w.grad.item(), 'w', w.item())
    w.grad.zero_()
```
